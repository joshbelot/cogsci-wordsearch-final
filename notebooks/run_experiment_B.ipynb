# Experiment B Notebook

This notebook runs the part B experiment and saves the data. Part A is puzzle word scaling. This notebook will also generate new puzzles if they don't exist already. 

## Imports

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
```

```python
import pandas as pd
import matplotlib
from tqdm import tqdm
from wordsearch.puzzle_sets import get_part_b
from wordsearch.strategies import (
    BruteForceStrategy,
    OrderedSearchStrategy,
    UncommonLetterSearchStrategy,
    RandomSampleStrategy,
    PatchSearchStrategy,
    RandomToOrderedStrategy
)
```

## Puzzle Setup
This will generate all the word searches needed if they don't already exist

```python
puzzles = get_part_b()
```

## Experiment Setup

```python
strategies = [
    BruteForceStrategy(),
    OrderedSearchStrategy(),
    UncommonLetterSearchStrategy(),
    RandomSampleStrategy(max_attempts=5000),
    PatchSearchStrategy(),
    RandomToOrderedStrategy(),
]
```

# Run Experiments and Collect Data

```python
rows = []

for condition, puzzle_list in puzzles.items():
    for i, puzzle in enumerate(tqdm(puzzle_list, desc=f"{condition:8s}", leave=True)):
        puzzle_id = f"B_{condition}_{i:04d}"
        word_lengths = [len(w) for w in puzzle.word_bank]
        avg_word_length = sum(word_lengths) / len(word_lengths)

        for strategy in strategies:
            result = strategy.solve(puzzle)

            rows.append({
                "puzzle_id":       puzzle_id,
                "strategy":        result.strategy_name,
                "condition":       condition,          # "short" / "medium" / "long"
                "size":            puzzle.size,        # always 30
                "word_count":      result.total_words,
                "avg_word_length": avg_word_length,
                "cells_examined":  result.cells_examined,
                "execution_time":  result.execution_time,
                "words_found":     result.words_found_count,
                "total_words":     result.total_words,
            })
```

```python
df_a = pd.DataFrame(rows)
print(f"\nTotal rows: {len(df_a):,}")
df_a.head()
```

```python
df_a.to_csv('../data/part_b/part_b_results.csv')
```
